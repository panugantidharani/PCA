```python
#import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.preprocessing import scale
```

```python
#import dataset
wine=pd.read_csv('wine (1).csv')
wine
```

```python
wine['Type'].value_counts()
```

```python
wine2=wine.iloc[:,1:]
wine2
```

```python
wine2.shape
```

```python
wine2.info()
```

```python
#converting data into numpy array
wine_array=wine2.values
wine_array
```

```python
#Normalizing the numerical data
wine_norm=scale(wine_array)
wine_norm
```

# PCA Implementatiom

```python
#Applying PCA Fit Transform to dataset
pca=PCA(n_components=13)

wine_pca=pca.fit_transform(wine_norm)
wine_pca
```

```python
#PCA components matrix or covariance Matrix
pca.components_
```

```python
#The amount of variance that each pca has
var=pca.explained_variance_ratio_
var
```

```python
#cumulative variance of each PcA
var1=np.cumsum(np.round(var,4)*100)
var1
```

```python
#variance plot for PCA components obtained
plt.plot(var1,color='yellow')
```

```python
#Final Dataframe
final_df=pd.concat([wine['Type'],pd.DataFrame(wine_pca[:,0:3],columns=['pc1','pc2','pc3'])])
final_df
```

```python
#visualisation of PCAs
fig=plt.figure(figsize=(16,12))
sns.scatterplot(data=final_df)
```

# checking with other clustering Algorithms

# 1.Hierarchial Clustering

```python
#import libraries
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import normalize
```

```python
#as we already have normalized data,create Dendrograms
plt.figure(figsize=(10,8))
dendrogram=sch.dendrogram(sch.linkage(wine_norm,'complete'))
```

```python
#create Clusters
hclusters=AgglomerativeClustering(n_clusters=3,affinity='euclidean',linkage='ward')
hclusters
```

```python
y=pd.DataFrame(hclusters.fit_predict(wine_norm),columns=['clustersid'])
y['clustersid'].value_counts()
```

```python
#Adding clusters to dataset
wine3=wine.copy()
wine3['clustersid']=hclusters.labels_
wine3
```

# 2.K-Means Clustering

```python
# Import Libraries
from sklearn.cluster import KMeans
```

```python
#using within-cluster sum-of-sqaures criteria
wcss=[]
for i in range (1,6):
        kmeans=KMeans(n_clusters=i,random_state=2)
        kmeans.fit(wine_norm)
        wcss.append(kmeans.inertia_)
```

```python
#plot K values vs wcss to get elbow graph
plt.plot(range(1,6),wcss)
plt.title('Elbow Graph')
plt.xlabel('Number of clusters')
plt.ylabel('wcss')
plt.show()
```

# Build Cluster algorithm using K=3

```python
#cluster algorithm using k=3
cluster3=KMeans(3,random_state=30).fit(wine_norm)
cluster3
```

```python
cluster3.labels_
```

```python
#assign clusters to the dataset
wine4=wine.copy()
wine4['cluster3id']=cluster3.labels_
wine4
```
